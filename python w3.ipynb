```python
## Linux
##sherrys-MacBook-Air:python_files Sherryliu$ mkdir week5/
##sherrys-MacBook-Air:python_files Sherryliu$ cd week5
import numpy as np
import pandas as pd
# import matplotlib.pyplot as plt
from IPython import get_ipython
mgc = get_ipython().magic
mgc(u'%matplotlib')
fil = pd.read_csv("~/Desktop/python_files/week5/train_titanic.csv")
#Q1
fil.info()
#Q3
fil.drop(1, inplace=True)  # Drop the row with index 1
#Q4
fil.drop('Embarked', inplace=True, axis=1)  # Drop the column called ‚ÄúEmbarked‚Äù
#Q5
fil.dropna(axis=0, inplace=True)  # Drop rows with null values
#Q6
fil['Fare'].hist()  # Plot fare histogram
#Q7
fil['Survived'].hist()  # Plot survived histogram
#Q8
print(np.corrcoef(fil['Fare'], fil['Survived']))  # Draw correlation matrix
# Q9 Interpret your results (the way you think might be
# interesting to explore in more details) and print your text as output:
print('Fare and Survive rate have very weak Pearson correlation')

# Advanced
#Q1
fil1 = pd.read_csv("~/Desktop/python_files/week5/train_titanic.csv")
fil1['doubledFare'] = fil1['Fare']*2  # create column doubledFar
# Q2 Fill null age values by mean(age):
fil1['Age'].fillna(fil1['Age'].mean(), inplace=True)
# Q3 Scatter plot age to fare and interpret:
fil1.plot.scatter(x='Age', y='Fare', c='DarkBlue')
print("Mid-age group tend to have higher fare")

# Q4 Present using groupby the count, mean, and
# standard deviation of survived column:
print(fil1.groupby('Survived').count())
print(fil1.groupby('Survived').mean())
print(fil1.groupby('Survived').std())

# Q5 Find top 3 correlated attributes to survive:
print(fil1.corr())
print("We see that Pclass, Fare/doubledFare, and Parch are the three most correlated attributes with Survived")
```
